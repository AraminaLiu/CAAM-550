# Problem1

### (a)

```python
def NewtonInterpolate(x, f):
    m = len(x)
    n = len(f)
    if m != n:
        return "vector x and f should have the same size."
    
    x = np.copy(x)
    a = np.copy(f)
    
    for k in range(1, m):
        a[k:m] = (a[k:m] - a[k-1])/(x[k:m] - x[k-1])
    return a
```

### (b)

```python
def HornerScheme(a, x, x_):
    n = len(a)
    n_ = len(x_)
    res = []
    
    for i in range(n_):
        p = a[n-1]
        for j in range(n-2, -1, -1):
            p = p*(x_[i] - x[j]) + a[j]
        res.append(p)
    
    return res
```

### (c)

```python
def update_interpolating_poly(a, x, xnew, fnew):
    n = len(x)
    
    for i in range(n):
        if xnew == x[i]:
            raise ValueError("You should give another input value different from element in x!")
    
    res = fnew - a[0]
    mul = 1
    
    for i in range(n-1):
        mul = mul * (xnew-x[i])
        res = res - mul*(a[i+1])
        
    a_ = res / (mul * (xnew - x[-1]))
    anew = np.append(a, a_)
    return anew
```

```python
def inverse_interpolation(x,f,k):
    res = np.zeros((k,2))
    x_ = np.array([0])
    
    for i in range(k):
        idx = i+1
        a = NewtonInterpolate(f[:idx],x[:idx])
        poly = HornerScheme(a, x[:idx], x_)
        res[i][0] = i+1
        res[i][1] = poly[0]
    
    return res
```

```python
def generate_xy(n):
    x = np.zeros(n)
    y = np.zeros(n)
    
    for i in range(n):
        x[i] = -5 + 10 * i / 4
        y[i] = -5 + 10 * i / 4
    return x,y
```

```python
def func(x,y):
    n=len(x)
    m=len(y)
    z = np.zeros((n,m))
    
    for i in range(n):
        for j in range(m):
            z[i][j] = (1/(1+x[i]**2)) * (1/(1+y[j]**2))
    return z
```

```python
def lagrange(X,Y,Z,x,y):
    n = len(X)
    m = len(Y)
    
    n_ = len(x)
    m_ = len(y)
    z_ = np.zeros((n_,m_))
    
    for idx1 in range(n_):
        for idx2 in range(m_):
            z = 0
            
            for i in range(n):
                for j in range(m):
                    p, q = 1, 1
                    for k in range(n):
                        if k == i:continue
                        p = p * (x[idx1]-X[k])
                        q = q * (X[i] - X[k])
                    for l in range(m):
                        if l == j:continue
                        p = p * (y[idx2]-Y[l])
                        q = q * (Y[j] - Y[l])
                    z = z + Z[i][j]*p/q
            z_[idx1][idx2] = z
            
    return z_
```

```python
X,Y = generate_xy(5)
Z = func(X,Y)
x = np.linspace(-5,5,100)
y = np.linspace(-5,5,100)
z = lagrange(X,Y,Z,x,y)
```

```python
def newton_2d(X,Y,F,x,y):
    n = len(X)
    m = len(Y)
    
    a = np.zeros((n,m))

    for i in range(n):
        for j in range(m):
            a[i][j] = F[i][j]
            
    for i in range(n):
        for q in range(1,m):
            for j in range(q,m):
                a[i][j] = (a[i][j] - a[i][q-1]) / (y[j] - y[q-1])
    
    for j in range(m):
        for p in range(1,n):
            for i in range(p,n):
                a[i][j] = (a[i][j] - a[p-1][j]) / (x[i] - x[p-1])

    n_ = len(x)
    m_ = len(y)
    p = a[n-1][m-1]
    M = np.zeros((n_,m_))
    res = np.zeros((n_,m_))
    
    for j in range(n_):
        for i in range(n-2, -1, -1):
            M[i][j] = M[i][j] * (x[j] - X[i]) + a[i][j]
            
    for j in range(m_):
        for i in range(m-2, -1, -1):
            res[i][j] = res[i][j] * (y[j] - Y[i]) + M[i][j]
            
    return res
```

```python
def PiecewiseLinear(f):
    m = len(f)
    h = 1/(m-1)
    n = 1/h
    
    c0 = np.zeros(int(n))
    c1 = np.zeros(int(n))
    
    for i in range(1,int(n)):
        c1[i-1] = (f[i] - f[i-1])/h
        c0[i-1] = f[i] - c1[i-1]*i*h
    return (c0, c1)
```

```python
def func(x):
    n = len(x)
    f = np.zeros(n)
    for i in range(n):
        f[i] = math.sin(3*x[i])
    return f
```

### (b)

```python
def EvalPiecewiseLinear(c0,c1,xnew):
    nc = len(c0)
    h = 1/nc
    nnew = len(xnew)
    
    idx = (np.trunc(xnew/h)).astype(int)
    idx = np.where(idx == nc, nc-1, idx)
    idx = list(map(int, idx))
    
    c1_new = c1[idx]
    c0_new = c0[idx]
    y = c0_new + c1_new * xnew
    
    return y
```

### (d)

##### $||p-f||_{L^2([0,1])} = ( \int_{0}^{1} |p(x)-f(x)|^2 \,dx )^{\frac{1}{2}}$
##### Apply Simpson's Rule, $ \int_{0}^{1} |p(x)-f(x)|^2 \,dx \approx \frac{1-0}{6} (|p(0)-f(0)|^2 + 4|p(0.5)-f(0.5)|^2 + |p(1)-f(1)|^2)$
##### In this problem, I will choose 500 points in the interval [0,1] and apply simpson's rule in every subinterval for an accurate estimate.

```python
def error_cal(func):
    err_seq = []
    err_bond = []
    for n in np.arange(5,101):
        err_bond.append(1/n)
        x = np.linspace(0,1,n+1)
        f = func(x)
        c0, c1 = PiecewiseLinear(f)
        x_new = np.linspace(0,1,500)
        p = EvalPiecewiseLinear(c0,c1,x_new)
        diff = p - func(x_new)
        # using Simpson's rule to approximate error
        err_aprox = 0
        for i in range(0,len(diff)-3,2):
            err_aprox_i = x_new[2]/6*(diff[i]**2 + 4*diff[i+1]**2 + diff[i+2]**2)
            err_aprox += err_aprox_i
        err_seq.append(math.sqrt(err_aprox))
    # plot the error sequence
    plt.loglog(np.arange(5,101),err_seq, label = 'Approximate error')
    plt.loglog(np.arange(5,101),err_bond, label = 'Error bound')
    plt.xlabel('n')
    plt.ylabel('$L_{aprox}^2 error$')
    plt.legend()
    plt.show()
```

```python
def Lerr2(p,f):
    diff = p-f
    n = len(diff)
    lerr2 = 0
    
    for i in range(0,n-3,2):
        lerr = math.sqrt((1/6)*((abs(diff[i])**2) + 4*(abs(diff[i+1])**2) + (abs(diff[i+2])**2)))
        lerr2 = lerr2 + lerr
    return lerr2
```

```python
err = []

for n in range(5,101):
    x = np.linspace(0,1,n+1)
    c0, c1 = PiecewiseLinear(func1(x))
    xnew = np.linspace(0,1,501)
    p = EvalPiecewiseLinear(c0, c1, xnew)
    ftrue = func1(xnew)
    lerr = Lerr(p, ftrue)
    err.append(lerr)
    
n = range(5,101)
fig = plt.figure()
plt.loglog(n, err, label="$L^{\infty}$")

plt.xlabel('$n$')
plt.legend()
plt.show()
```
